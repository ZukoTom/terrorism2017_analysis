```python
#on réaligne certaines lignes décallées des colonnes
matching_rows = d.natlty1.str.contains('[a-zA-Z]', regex= True, na=False)
while True in matching_rows.to_numpy(): 
    d.loc[matching_rows, 'natlty1':] = d.loc[matching_rows, 'natlty1':].shift(-1, axis = 1)
    matching_rows = d.natlty1.str.contains('[a-zA-Z]', regex= True, na=False)
matching_rows.unique()
```

```python
matching_rows2 = d.guncertain1.str.contains('[a-zA-Z]', regex= True, na=False)
while True in matching_rows2.to_numpy(): 
    d.loc[matching_rows2, 'guncertain1':] = d.loc[matching_rows2, 'guncertain1':].shift(-1, axis = 1)
    matching_rows2 = d.guncertain1.str.contains('[a-zA-Z]', regex= True, na=False)
matching_rows2.unique()
```

```python
matching_rows3 = d.INT_LOG.str.contains('[a-zA-Z]', regex= True, na=False)
while True in matching_rows3.to_numpy(): 
    d.loc[matching_rows3, 'INT_LOG':] = d.loc[matching_rows3, 'INT_LOG':].shift(-1, axis = 1)
    matching_rows3 = d.INT_LOG.str.contains('[a-zA-Z]', regex= True, na=False)
matching_rows3.unique()
```

```python
#on selectionne les var pertinentes pour notre étude
d2 = d[['country_txt','region_txt','latitude','longitude','doubtterr','multiple','success',\
        'suicide','attacktype1_txt','targtype1_txt','natlty1_txt','nperpcap','weaptype1_txt',\
        'nkill','nkillter','nwound','INT_LOG']]
```

```python
#on séléctionne un sous échantillon
dpca = d[['doubtterr','multiple','success','suicide','nperpcap','nkill','nkillter','nwound']]
```

```python
#on exclue les lignes avec des nan (donnees manquantes)
d2 = d2.dropna()
dpca = dpca.dropna()
```

```python
#on verifie qu'il ne reste plus de nan
for i in range(len(d2.columns)):
    if (d2[d2.columns[i]].isnull().sum() != 0):#.values):
        print(d2[d2.columns[i]].isnull().sum())
    else:
        print(f"plus de nan (total de nan colonne {i} = {d2[d2.columns[i]].isnull().sum()})")
```

```python
for i in range(len(d2.columns)):
    print(d2.columns[i], d2[d2.columns[i]].unique())
```

```python
#PARTIE ACP
#prepa donnees on ne prends que les var quant
X = dpca.select_dtypes('number').values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
#centrage réduction
X_scaled = scaler.transform(X)
```

```python
#calcul composantes
from sklearn.decomposition import PCA
pca = PCA()#n_components=3?)
p = pca.fit(X_scaled)
x = np.arange(X.shape[1])
print("Pourcentage de variance expliquée : ")
plt.bar(x, p.explained_variance_ratio_)
print("Composantes principales : ")
#les coordonnées des PC dans le repère des vecteurs variables de depart
print(p.components_)
```

```python
from mpl_toolkits.mplot3d import Axes3D
comp = p.transform(X_scaled)
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection="3d", elev=30, azim=15)
x,y,z = comp[:,0], comp[:,1], comp[:,2]
ax.scatter(x,y,z, marker=".")
#ax.text(x,y,z,noms[i])
plt.show()
```
